Bai 1

```
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
data = pd.read_csv('./Diemsv.csv')
y = data.iloc[:, -1]

data['Toan'],_=pd.factorize(data['Toan'])
data['Ly'],_=pd.factorize(data['Ly'])
data['Hoa'],_=pd.factorize(data['Hoa'])
X= data.loc[:,['Toan','Ly','Hoa']].values
clf = DecisionTreeClassifier(max_depth = 6, min_samples_split = 2, criterion='entropy')
clf.fit(X,y)
tree.plot_tree(clf.fit(X,y))
print(clf.predict(X))
```

```
from sklearn.ensemble import RandomForestClassifier

	rf = RandomForestClassifier(n_estimators=100)
	rf.fit(X,y)
 print(rf.predict(X))
```
